**Memory:**

Memory is to store the raw data or information in the computer. In
computers there would be temporary memory storage and permanent memory
storage just like how humans store their information in the short term
and long-term memory.

Memory is of three types:

1\. Primary memory

2\. Secondary memory

3\. Cache memory

1.  **Primary memory:**

We can call primary memory as main memory or semiconductor memory.
Primary memory uses semiconductor technology \[ note: semiconductor is a
material that conducts electricity more than the insulator but less than
the conductor. semiconductors can be found in thousands of products such
as computers, smartphones, memory, microprocessors, etc.\].

Primary memory is **volatile or non-volatile** in nature \[note :
volatile means as soon as the computer is switched off, the information
stored in the main memory would be erased.\]

**Volatile memory is a temporary data storage that requires power to
keep the data.**

**Non-volatile memory is a permanent data storage that does not require
power to keep the data.**

Types of primary memory:

> Based on **Volatile**:

I. RAM

> II\. Cache memory
>
> Based on **Non-volatile**:

III\. ROM

1.  **RAM (Random Access Memory)** is a volatile memory. The data or
    instructions in the RAM is directly accessible by the microprocessor
    at a faster speed. RAM is used for booting, running applications,
    loading files, performing calculations and so on. **Without RAM a
    computer cannot run.**

> They are two types of RAM:
>
> 1.SRAM
>
> 2.DRAM
>
> **1.SRAM (static RAM)** used for faster, energy efficient and small.
> Static RAM uses Transitions and circuits. SRAM is faster and more
> expensive than DRAM and is used in applications where speed and
> efficiency are critical and provides less access time. It uses Flip
> Flop to store information.
>
> **2.DRAM (Dynamic RAM)** uses Capacitors and Transistors to store
> data, Less cost. The name refers that it should be refreshed
> periodically. laptops, mobile devices use DRAM

**II. Cache memory:** It is high-speed memory used by the CPU to store
often used data and instructions to improve system performance.

**III. ROM (Read Only Memory)** is a non-volatile memory. It is used for
long-term storage of data that needs to be retained even when power is
turned off or lost. It can be used to store system firmware, BIOS, or
other critical data that needs to be retained even when the power is
turned off.

> Four types of ROM:
>
> 1\. MROM
>
> 2.PROM
>
> 3.EPROM
>
> 4.EEPROM
>
> 5.Flash Memory

1.  **MROM (Mask ROM):** It stores the information permanently with a
    set of instructions or data during the manufacturing process. Once
    the programming is done it cannot be changed or updated.

2.  **PROM (Programmable ROM):** It allows the user to program the
    memory after it is manufactured.

3.  **EPROM (Erasable PROM):** It is programmed and erased multiple
    times using ultraviolet light.

4.  **EEPROM (Electrically EPROM):** The data needs to be retrieved even
    when power is removed or where the frequency updates are required.

5.  **Flash memory:** Flash memory is special because it can be
    programmed and erased multiple times. It works by using tiny
    electrical charges to store data in small memory cells. These memory
    cells are organized into blocks, which can be erased and
    reprogrammed when needed.

2\. **Secondary memory:**

We can call it auxiliary memory or backup memory. It is non-volatile
memory used to store large data permanently. It is slower than Primary
memory. CPU cannot access directly from secondary memory. Without
secondary memory computers can run.

types of secondary memory:

1\. Magnetic Tapes.

2.Magnetic disks.

3\. Optical disks.

1.  **Magnetic Tapes:** Sequential access storage media means data
    stored and retrieved in linear order. This helps them to be useful
    for storing enormous amounts of data. It is used for backup and
    archival purposes.

2.  **Magnetic disks:** Random access storage. It is used as a primary
    storage device and provides fast access. It can be either HDDs or
    SSDs. HDDs use spinning disks coated with magnetic material to store
    the data (stores on sectors and tracks) while SSDs use NAND flash
    memory chip to store the data. SSDs do not have moving parts which
    makes them faster and more reliable than HDDs.

3.  **Optical disks:** laser-based storage medium: CDs and DVDs.

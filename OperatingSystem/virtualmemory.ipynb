Virtual Memory is a software program not a physical device. It is
enabled by default when the OS (Operating System) is installed on your
computer. Virtual memory role is to extend the RAM (Random Access
Memory) space to work more efficiently. Virtual memory uses some portion
of Hard Drive to temporarily store data that is not currently being used
by running programs when the physical memory (RAM) is running low. When
the data is needed again, it is swapped back into physical memory from
the virtual memory storage area (e.g., pagefile or swap space).

Physical Memory does not extend to the hard drive or ROM (read-only
memory), as those are separate components with different functions in a
computer. It extends only the RAM (physical memory).

Windows operating systems use a virtual memory system called the
“pagefile”, which is a file located on the hard drive that is used to
supplement physical memory (RAM) when needed. In MacOS, the virtual
memory system is called “swap”. On Unix/Linux systems, the virtual
memory storage area is called “swap space”.

Virtual memory is a way for a computer to efficiently manage its
available resources and ensure that it can handle multiple tasks at once
without crashing or slowing down.

**Why do we need to use Virtual Memory?**

Virtual memory is used because it allows a computer to run more programs
than the amount of physical memory (RAM) available in the system. By
using virtual memory, a computer can temporarily store data on the hard
drive when it runs out of physical memory, freeing up space in RAM for
other tasks. This allows the computer to run more programs than it would
otherwise be able to and prevents crashes caused by running out of
memory.

Virtual memory also allows for the efficient sharing of memory resources
between different programs running on the computer. This means that
multiple programs can use the same physical memory without interfering
with each other, which enhances the system’s overall performance.

**What happens if we do not use Virtual Memory?**

If your computer does not have virtual memory, it would run out of
physical memory (RAM) as soon as the number of programs running exceeded
the amount of available memory. This would cause the computer to become
extremely slow or even crash and severely limit the number of programs
which run simultaneously. If a large amount of memory needs to run
properly, and without virtual memory we cannot be able to run multiple
programs.

**How does virtual memory work?**

When a program is run, it is divided into smaller parts called
**pages**. These pages are loaded into physical memory (RAM) as needed.

However, when the physical memory is running low, the operating system
will move some pages out of physical memory and store them on the hard
drive. This process is called **“paging out” or “swapping out”** and it
frees up space in physical memory for other programs to use.

When the program needs to access a page that has been paged out, the
operating system will “**page in” or “swap in**” the necessary page from
the hard drive back into physical memory.

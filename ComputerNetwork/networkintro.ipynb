**Computer Networks:**

Computer network is a group of computers, servers, and other devices
connected to each other (why do they have to connect?) to share
resources and communicate with each other. The resources might be
printers, internet connections, files and so on. How can a computer
communicate? Through a communication protocol (protocol is just a set of
rules to follow, to communicate with other devices). When computers are
connected, they can share things like pictures, videos, and messages
with each other.

In a quite effortless way, Assume you and your friend work in different
companies as a software engineer. If you would like to communicate with
him, you can use phone call, video call or messaging app (these are the
resources) with communication protocol (like you must know your friend
free time to communicate – rules).

**Why this computer network is important for us:**

To transfer the data between devices and system. To get internet access,
which is essential for accessing information, communication, and data
transfer.

**What happens if we** **don’t use the Computer Network technology:**

If we don’t use the Computer Network concepts, we just rely on the
traditional methods for communication like sending messages through the
letters and rely on physical storage devices such as hard drives or USBs
to transfer the data between computers and so on.

**Types of networks:**

1.**Local Area Network (LAN):** - devices would be connected in a small
area most commonly using wired connections such as EHTERNET cables used
to form the LAN network.

A LAN can be connected to a [wide area
network](https://en.wikipedia.org/wiki/Wide_area_network) (WAN) using a
[router](https://en.wikipedia.org/wiki/Router_(computing)).

2\. **Wide Area Network (WAN):** - devices would be connected in a large
area using both wire and wireless connections. Wireless connections for
WANs typically use technologies such as microwave, satellite, or
cellular networks. Wired connections for WANs such as Fiber optic
cables, copper wires, coaxial cables, and leased lines.

3\. **Metropolitan Area Network (MAN):** - A MAN is a network that
connects devices within a metropolitan area, such as a city.

4\. **Wireless Local Area Network (WLAN)**: A WLAN is a LAN that uses
wireless technology, such as Wi-Fi, to connect devices together without
using cables.

5\. **Storage Area Network (SAN)**: A SAN is a specialized network that
is used to connect storage devices, such as disk arrays or tape
libraries, to servers.

6.**Virtual Private Network (VPN)**: A VPN is a network that uses the
public internet to create a secure, encrypted connection between remote
users and a private network.

7.**Peer-to-Peer Network (P2P):** A P2P network is a decentralized
network where all devices can act as both a client and a server,
allowing for direct communication and file sharing between them.

**Ethernet** is the most reliable and fastest network for data transfer
over short distances, while **cellular** networks are the best choice
for long-distance data transfer. **Wi-Fi and Bluetooth** are both
convenient for mobile devices and can be useful for transferring lesser
amounts of data.

Which is best for data transmission?

Ethernet networks are generally considered the best choice for data
transfer within a local area network (LAN), such as within a home or
office building. Ethernet is a wired network that provides reliable and
fast data transfer, with speeds ranging from 10 Mbps to 10 Gbps
depending on the equipment used. Ethernet is also more secure than
wireless networks, as it **requires physical access to the network cable
to intercept data.**

Cellular networks, on the other hand, use **radio waves to** **transmit
data** over a wireless network and can provide data transfer speeds
ranging from 3G to 5G depending on the network and device. Cellular
networks are convenient because they allow users to connect from
anywhere, but they can be less reliable and slower than Ethernet
networks, especially in areas with poor signal strength or network
congestion.

**Wired connections** generally offer higher speeds, greater
reliability, and lower latency, while **wireless connections** offer
greater mobility and flexibility. However, wireless connections can be
affected by factors such as distance, signal interference, and network
congestion.
